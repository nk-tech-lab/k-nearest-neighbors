```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# create the artificial dataset 

# generates a 500 sample data set with 2 features, 5 centers, and a cluster standard deviation of 1.5
x, y = make_blobs(n_samples=500, n_features=2, centers=5, cluster_std=1.5, random_state=4)
```

```python
# visualize the data using matplotlib

sns.set_theme('notebook')
plt.figure(figsize=(10, 10))
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, edgecolors='black', marker='*')
plt.title('Artificial Data Set Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# split the data into training and testing sets
# 80% of the data will be used for training and 20% will be used for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
# create the KNN model
knn5 = KNeighborsClassifier(n_neighbors=5)
knn10 = KNeighborsClassifier(n_neighbors=10)
knn1 = KNeighborsClassifier(n_neighbors=1)
knn100 = KNeighborsClassifier(n_neighbors=100)
```

```python
# train the model
knn5.fit(x_train, y_train)
knn10.fit(x_train, y_train)
knn1.fit(x_train, y_train)
knn100.fit(x_train, y_train)
```

```python
# predict on actual values
y_pred5 = knn5.predict(x_test)
y_pred10 = knn10.predict(x_test)
y_pred1 = knn1.predict(x_test)
y_pred100 = knn100.predict(x_test)
```

```python
# accuracy score
accuracy5 = accuracy_score(y_test, y_pred5)
accuracy10 = accuracy_score(y_test, y_pred10)
accuracy1 = accuracy_score(y_test, y_pred1)
accuracy100 = accuracy_score(y_test, y_pred100)
print(f'Accuracy for 5 neighbors: {accuracy5}')
print(f'Accuracy for 10 neighbors: {accuracy10}')
print(f'Accuracy for 1 neighbor: {accuracy1}')
print(f'Accuracy for 100 neighbors: {accuracy100}')
```

```python
# visualize the decision boundaries
# sns.set_theme('notebook')
plt.figure(figsize=(20, 20))
# plt.title('KNN Decision Boundaries', fontsize=20)

# plot k = 5
plt.subplot(2, 2, 2)
plt.title('Predicted values with k=5')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, edgecolors='black', marker='*')
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred5, s=100, edgecolors='black', marker='o')

# plot k = 10
plt.subplot(2, 2, 3)
plt.title('Predicted values with k=10')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, edgecolors='black', marker='*')
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred10, s=100, edgecolors='black', marker='o')

# plot k = 1
plt.subplot(2, 2, 1)
plt.title('Predicted values with k=1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, edgecolors='black', marker='*')
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred1, s=100, edgecolors='black', marker='o')

# plot k = 100
plt.subplot(2, 2, 4)
plt.title('Predicted values with k=100')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.scatter(x[:, 0], x[:, 1], c=y, s=100, edgecolors='black', marker='*')
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred100, s=100, edgecolors='black', marker='o')
plt.show()
```
